##### How to check whether triangle forms or not
Sum of any two sides should be grater than third

```python
from dataclasses import dataclass
```

```python
@dataclass

class Triangle:
    a:int|float
    b:int|float
    c:int|float

    def is_triangle(self) ->bool:
        if (
            (self.a + self.b >self.c)
            and (self.b + self.c >self.a)
            and (self.a + self.c >self.b)
        ):
            return True
        else:
            return False
        
    def __post_init__(self):
        errors = []

        if not isinstance(self.a, (int, float)) or self.a <= 0:
            errors.append("Side a should be a positive int or float")
        if not isinstance(self.b, (int, float)) or self.b <=0:
            errors.append("Side b should be a positive int or float")
        if not isinstance(self.c, (int, float)) or self.c <=0:
            errors.append("Side c should be a positive int or float")
        if not self.is_triangle():
            errors.append("Given sides do not form a triangle")

        if errors:
            raise ValueError("\n".join(errors))
        
    def perimeter(self):
        return self.a + self.b + self.c 
    
    def area(self):
        s = self.perimeter()/2
        a = (s*(s-self.a)*(s-self.b)*(s-self.c))**(1/2)
        return a
```

### Bank account example

```python
from dataclasses import dataclass
```

```python
@dataclass
class account:
    ac_no: int
    name: str
    balance: float

    def show_current_balance(self):
        print(f"A/C no-{self.ac_no}, Name : {self.name}, Current balance : {self.balance}")

    def deposit(self,amt:float):
        if amt <= 0:
            raise ValueError("Amount to be deposited cannot be negative or zero")
        self.balance = self.balance + amt

    def withdraw(self,amt:float):
        if amt <=0:
            raise ValueError("Amount withdrawn cannot be negative or zero")
        if amt >= self.balance:
            raise ValueError("Inseffecient Balance")
        
        self.balance = self.balance - amt

    def tranfer(self,amt:float, ac):
        self.withdraw(amt)
        ac.deposit(amt)
```

### Polymorphism in classes

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken language in India")
```

```python

class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
c2 = USA()
c3 = France()
```

```python
c1.capital()
```

```python
c2.capital()
```

```python
c1.language()
```

```python
c2.capital()
```

```python
for i in (c1,c2,c3):
    print(f"Type : {type(i)}")
    i.capital()
    i.language()
    print("\n============================\n")
```

##### Calculating area and perimeter for different shapes

```python
@dataclass

class rectangle:
    length: int| float
    width: int| float

    def perimeter(self):
        return 2 * (self.length + self.width)
    def area(self):
        return self.length * self.width
```

```python
import math
```

```python
math.pi
```

```python
@dataclass

class circle:
    radius:int | float

    def perimeter(self):
        return 2 *math.pi * self.radius
    def area(self):
        return math.pi * (self.radius**2)
```

```python
s1 = Triangle(3,4,6)
s2 = circle(7)
s3 = rectangle(20,10)
s4 = circle(10)
s5 = Triangle(12,13,14)
```

```python
for i in (s1,s2,s3,s4,s5):
    print(i)
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("\n================================\n")
```
